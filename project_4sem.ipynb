```python
df['house_ownership']= df['house_ownership'].map({'rented':0,'owned':1,'norent_noown':2})
```

```python
df['car_ownership']= df['car_ownership'].map({'no':0,'yes':1})
```

```python
df['married']= df['married'].map({'single':0,'married':1})
```

```python
### split dataset into indepenent and dependent variable
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```
